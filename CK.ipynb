```python
import cv2
import glob

import os  
import csv
import numpy as np
import random
import dlib
from skimage import io
```

```python
#Define 7 types of emotions, corresponding to the 7 types in CK+ dataset
emotions = ["anger", "contempt", "disgust", "fear", "happy", "sadness", "surprise"]
rootpath1 = "/Users/scott/Documents/2017Fall/UROP1100/Emotion"

#Extract label information from CK+ dataset folder
def listdir(path):  
    list_dir = os.listdir(path)
    return list_dir
label = []
label_file = []
list_dir = listdir(rootpath1)
a = 0
list_dir.sort()
for i in range(1,len(list_dir)):
    if list_dir[i]!=".DS_Store":
        path_1 = rootpath1 +"/"+list_dir[i]
        list_dir_1 = listdir(path_1)
        #print(list_dir_1)
        list_dir_1.sort
        for j in range(0,len(list_dir_1)):
            if list_dir_1[j]!=".DS_Store":
                path_2 = path_1+"/"+list_dir_1[j]
                list_dir_2 = listdir(path_2)
                list_dir_2.sort
                if(list_dir_2!=[]):
                    #print(list_dir_2)
                    a = a+1
                    label_file.append(list_dir_2[0])
                    with open(path_2+"/"+list_dir_2[0]) as f:
                        content = f.readlines()
                        #print(content[0].rstrip('.0000000e+00\n'))
                        label.append(int(content[0].rstrip('.0000000e+00\n')))
print(label_file)
print(len(label))
```

```python
#Extract Landmark information from CK+ dataset folder
f1 = []
f2 = []
for i in range(len(label_file)):
    temp = label_file[i].rstrip('_emotion.txt')
    print(temp)
    f2.append(temp)
    temp = temp.split('_')
    f1.append(temp)
    print(f1[i])
rootpath2 = "/Users/scott/Documents/2017Fall/UROP1100/Landmarks"
X1 = []
X2 = []
landmark_file = []
landmark_file.append([])
landmark_file.append([])
add = 0
for i in range(len(label_file)):
    path = rootpath2 +"/" + f1[i][0] + "/" + f1[i][1]
    landmark_file[0].append(path+"/"+f1[i][0]+"_"+f1[i][1]+"_00000001_landmarks.txt")
    landmark_file[1].append(path+"/"+f2[i]+ "_landmarks.txt")
    with open(landmark_file[0][i]) as f:
        contents = f.readlines()
        tmpData = []
        for j in range(len(contents)):
            content = contents[j].rstrip('\n')
            content = content.split(' ')
            for k in range(len(content)):
                if(content[k]!='' and content[k]!='\n'):
                     tmpData.append(float(content[k]))
        X1.append(tmpData)
    with open(landmark_file[1][i]) as f:
        contents = f.readlines()
        tmpData = []
        for j in range(len(contents)):
            content = contents[j].rstrip('\n')
            content = content.split(' ')
            for k in range(len(content)):
                if(content[k]!='' and content[k]!='\n'):
                    tmpData.append(float(content[k]))
        X2.append(tmpData)
```

```python
#Preprocess the data
X1 = np.array(X1,dtype = np.float)
X2 = np.array(X2,dtype = np.float)
X_1 = np.zeros([327,68,2])
X_2 = np.zeros([327,68,2])
for i in range(327):
    for j in range(68):
        for k in range(2):
            X_1[i,j,k] = X1[i,j*2+k]
            X_2[i,j,k] = X2[i,j*2+k]
label = np.array(label,dtype=np.int )
X1 = X_1
X2 = X_2
```

```python
#Preprecess the data to the correct dimension
X = np.sqrt(np.sum(np.square(X1-X2),axis = 2))
X = np.sum((X1-X2),axis=2)
print(X.shape)
```

```python
from sklearn import preprocessing
X_norm = preprocessing.scale(X)
print(X_norm[0])
```

```python
#Perform PCA on the data
from sklearn.decomposition import PCA
pca = PCA(n_components='mle')
pca.fit(X_norm)
print pca.explained_variance_ratio_
print pca.explained_variance_
X_new = pca.transform(X_norm)
print(X_new.shape)
print(X_new)
```

```python
from sklearn import svm
from sklearn import cross_validation
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
```

```python
#Using liniear SVM and show the corresponding confusion matrix
clf = svm.SVC()
clf.fit(X_new[0:240], label[0:240])
result = clf.predict(X_new[240:])
num_correct=(np.sum(result==label[240:]))
accuracy = float(num_correct)/result.shape[0]
print('Got %d / %d correct => accuracy: %f' % (num_correct,result.shape[0],accuracy))
conf_mat = confusion_matrix(label[240:],result)
print(conf_mat)
```

```python
#Cross Validation resulte from linear SVM
clf = svm.SVC(kernel='linear', C=1)
scores = cross_validation.cross_val_score(clf, X_new, label, cv=5)
print(scores)
print(np.mean(scores))
```

```python
#Cross Valiation result Using Two layer neural network
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(20, 7), random_state=1)
scores = cross_validation.cross_val_score(clf, X_new, label, cv=5)
print(scores)
print(np.mean(scores))
```

```python
#Confusion matrix
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
y_pred = cross_val_predict(clf,X_new,label,cv=5)
conf_mat = confusion_matrix(label,y_pred)
print(conf_mat)
```
