```python
import csv
import numpy as np
import dlib
from skimage import io
import matplotlib.pyplot as plt
import cv2
import os  
import random
from sklearn import svm
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.neural_network import MLPClassifier
```

```python
a = 0;
datasize = [4000,10000,20000]
datascale = datasize[2]
img = np.zeros([48,48,datascale])
label = np.zeros([datascale,],dtype=np.uint8)
n = 60
with open("/Users/scott/Downloads/fer2013/fer2013.csv") as csvfile:
    reader = [each for each in csv.DictReader(csvfile)]
for row in reader:
    #print(row["emotion"],row["pixels"])
    pixelstring = row["pixels"].split(" ")
    emotionstring = row["emotion"]
    emotion = int(emotionstring)
    pixels = map(int,pixelstring)
    #print(pixels)
    a += 1
    imgs = np.array(pixels,dtype=np.uint8)
    img[:,:,a-1]= np.resize(imgs,(48,48))
    label[a-1,] = emotion
    #plt.imshow(imgs.reshape(48,48),cmap='gray')
    #plt.show()
    if(a == img.shape[2]):
        break
for i in range(10):
    plt.imshow(img[:,:,n+i],cmap ="gray")
    plt.show()
 #print(imgs)
```

```python
predictor_path = "/Users/scott/Downloads/shape_predictor_68_face_landmarks.dat"
cascade_path = ["/usr/local/Cellar/opencv/3.3.1_1/share/OpenCV/haarcascades/haarcascade_frontalface_default.xml",
                "/usr/local/Cellar/opencv/3.3.1_1/share/OpenCV/haarcascades/haarcascade_frontalface_alt.xml",
                "/usr/local/Cellar/opencv/3.3.1_1/share/OpenCV/haarcascades/haarcascade_frontalface_alt2.xml",
               "/usr/local/Cellar/opencv/3.3.1_1/share/OpenCV/haarcascades/haarcascade_frontalface_alt_tree.xml"]
features = np.zeros([12,2,datascale])
labels = np.zeros([datascale,1])
faceCascade = cv2.CascadeClassifier(cascade_path[1])
predictor = dlib.shape_predictor(predictor_path)
recog_num = 0
for i in range(datascale):
    #image = io.imread("/Users/scott/Downloads/cohn-kanade-images/S005/001/S005_001_00000001.png")#np.array(pixels,dtype=np.uint8)
    image = np.array(img[:,:,i],dtype=np.uint8)
    #win = dlib.image_window()
    #win.set_image(image)
    faces = faceCascade.detectMultiScale(  
       image,  
       scaleFactor=1.1,  
       minNeighbors=1,  
       minSize=(20, 20),  
       flags=cv2.CASCADE_SCALE_IMAGE  
    ) 
    #print("Found {0} faces!".format(len(faces))) 
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    dlib_rect = dlib.rectangle(int(x), int(y), int(x + w), int(y + h)) 
    shape = predictor(image, dlib_rect)
    allshape = shape.parts()
    if (len(faces)==1):
        features[0,:,recog_num] = np.array([allshape[19].x-allshape[17].x,allshape[19].y-allshape[17].y])
        features[1,:,recog_num] = np.array([allshape[21].x-allshape[19].x,allshape[21].y-allshape[19].y])
        features[2,:,recog_num] = np.array([allshape[24].x-allshape[22].x,allshape[24].y-allshape[22].y])
        features[3,:,recog_num] = np.array([allshape[26].x-allshape[24].x,allshape[26].y-allshape[24].y])
        features[4,:,recog_num] = np.array([allshape[51].x-allshape[48].x,allshape[51].y-allshape[48].y])
        features[5,:,recog_num] = np.array([allshape[54].x-allshape[51].x,allshape[54].y-allshape[51].y])
        features[6,:,recog_num] = np.array([allshape[57].x-allshape[48].x,allshape[57].y-allshape[48].y])
        features[7,:,recog_num] = np.array([allshape[54].x-allshape[57].x,allshape[54].y-allshape[57].y])
        features[8,:,recog_num] = np.array([allshape[19].x-allshape[36].x,allshape[19].y-allshape[36].y])
        features[9,:,recog_num] = np.array([allshape[19].x-allshape[39].x,allshape[19].y-allshape[39].y])
        features[10,:,recog_num] = np.array([allshape[24].x-allshape[42].x,allshape[24].y-allshape[42].y])
        features[11,:,recog_num] = np.array([allshape[24].x-allshape[45].x,allshape[24].y-allshape[45].y])
        features[:,:,recog_num] = features[:,:,recog_num]/np.linalg.norm(features[:,:,recog_num])
        labels[recog_num,] = label[i,] 
        recog_num+=1
    #print(features[:,:,0])
    #print(allshape[51])
    #print(type(allshape[0].x))
    #win.add_overlay(shape)
print(recog_num)
training_data_raw = features[:,:,0:int(0.8*recog_num)]
training_label_raw = labels[0:int(0.8*recog_num),]
predicting_data_raw = features[:,:,int(0.8*recog_num):recog_num]
predicting_label_raw = labels[int(0.8*recog_num):recog_num,]
```

```python
pca=PCA(n_components=1)
newData_training = np.zeros([training_data_raw.shape[0],training_data_raw.shape[2]])
newData_predicting = np.zeros([predicting_data_raw.shape[0],predicting_data_raw.shape[2]])
for i in range(training_data_raw.shape[2]):
    newData_training[:,i]=(pca.fit_transform(training_data_raw[:,:,i])).reshape(-1)
for i in range(predicting_data_raw.shape[2]):
    newData_predicting[:,i]=(pca.fit_transform(predicting_data_raw[:,:,i])).reshape(-1)

training_data_process = np.reshape(training_data_raw,[24,-1]).T
predicting_data_process = np.reshape(predicting_data_raw,[24,-1]).T
print(predicting_data_process.shape)
print(predicting_label_raw.shape)
training_label_process = np.array(training_label_raw,dtype = np.uint8)
```

```python
#X = training_data_process
X = newData_training.T
Y = training_label_process.reshape(-1)
clf = svm.SVC()
clf.fit(X, Y)
```

```python
#result = clf.predict(predicting_data_process)
result = clf.predict(newData_predicting.T)
result = np.array(result,dtype = np.uint8)
predicting_label = np.array(predicting_label_raw.reshape(-1),dtype = np.uint8)
print result.shape
print predicting_label.shape
```

```python
num_correct=(np.sum(result==predicting_label))
accuracy = float(num_correct)/result.shape[0]
print('Got %d / %d correct => accuracy: %f' % (num_correct,result.shape[0],accuracy))
print(result[5])
print(predicting_label[5])
```

```python
X = training_data_process
Y = training_label_process.reshape(-1)
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(100, 7), random_state=1)
clf.fit(X, Y)
```

```python
result = clf.predict(predicting_data_process)
result = np.array(result,dtype = np.uint8)
predicting_label = np.array(predicting_label_raw.reshape(-1),dtype = np.uint8)

num_correct=(np.sum(result==predicting_label))
accuracy = float(num_correct)/result.shape[0]
print('Got %d / %d correct => accuracy: %f' % (num_correct,result.shape[0],accuracy))
print(result[5])
print(predicting_label[5])
```
